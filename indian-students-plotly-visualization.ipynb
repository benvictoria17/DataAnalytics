```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.figure_factory as ff
import plotly.graph_objects as go
import plotly.express as px
from plotly.offline import iplot

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
df = pd.read_csv("../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv")
```

```python
df = df.iloc[1:, :]
df.head()
```

```python
df = df[(df['Q3'] == 'India') & (df['Q5'] == 'Student')]
df.head()
```

```python
print(f"We have Data for {len(df)} Indian Students")
```

# Education

```python
temp_df = df["Q4"].value_counts().reset_index()

trace1 = go.Bar(
                x = ['Bachelor’s degree', 'Master’s degree', "Some college/university without bachelor's degree", 'Doctoral degree','I prefer not to answer','Professional degree','No formal education past high school'], #temp_df['index'],
                y = temp_df['Q4'],
                marker = dict(color = 'rgb(250,13,92)',
                              line=dict(color='rgb(0,0,0)',width=1.5)),
                text=temp_df['Q4'], textposition='outside')
layout = go.Layout(template= "plotly_dark",title = 'EDUCATION LEVEL OF INDIAN STUDENT KAGGLERS' , xaxis = dict(title = 'Education Level'), yaxis = dict(title = 'Count'), height=650)
fig = go.Figure(data = [trace1], layout = layout)
fig.show()
```

```python
def pie_plot(cnt_srs, title):
    labels=cnt_srs.index
    values=cnt_srs.values
    trace = go.Pie(labels=labels, 
                   values=values, 
                   title=title, 
                   hoverinfo='percent+value+label', 
                   textinfo='percent',
                   textposition='inside',
                   hole=0.6,
                   showlegend=True,
                   marker=dict(colors=plt.cm.viridis_r(np.linspace(0, 1, 28)),
                               line=dict(color='#000000',
                                         width=2),
                              )
                  )
    
    return trace
```

```python
trace = go.Pie(labels=['Bachelor’s degree', 'Master’s degree', 'Some college/university without bachelor’s degree',
                       'Doctoral degree', 'I prefer not to answer', "Professional degree", 'No formal education past high school'], 
               values=df['Q4'].value_counts().values, 
               title="Education Level Distribution", 
               hoverinfo='percent+value+label', 
               textinfo='percent',
               textposition='inside',
               hole=0.6,
               showlegend=True,
               marker=dict(colors=plt.cm.viridis_r(np.linspace(0, 1, 28)),
                           line=dict(color='#000000',
                                     width=2),
                          )
                  )
fig=go.Figure(data=[trace])
fig.show()
```

### More than 63% students are pursuing their Bachelor's Degree
