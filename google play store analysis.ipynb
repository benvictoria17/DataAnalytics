```python
import pandas as pd 
import numpy as np 
import plotly.express as px
import plotly.graph_objects as go 
import plotly.figure_factory as ff
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataVisualization/master/dataset/Google%20Play%20Store%20Apps/googleplaystore.csv")
```

```python
df.head()
```

```python
print("Number of Columns in this dataset has:", df.shape[1])
print("Number of Columns in this dataset has:", df.shape[0])
```

```python
df.describe(include = "object")
```

```python
df.isnull().sum()
```

```python
def info_data():
    data=pd.DataFrame(index=df.columns)
   
    data["Data_type"]=df.dtypes
    data["Null Count"]=df.isnull().sum()
    
    data["Ununique_count"]=df.nunique()
    return data
```

```python
info_data()
```

```python
df[df.Type.isnull()]
```

```python
df["Type"].fillna("Free", inplace=True)
```

```python
df.Type.isnull().sum()
```

```python
df[df["Content Rating"].isnull()]
```

```python
df.dropna(subset=["Content Rating"],inplace=True)
df["Content Rating"].isnull().sum()
```

```python
mode=df.Rating.mode()
mode
```

```python
df["Rating"].fillna(value=mode[0],inplace=True)
df.Rating.isnull().sum()
```

```python
df.dropna(subset=["Current Ver"],inplace=True)
df["Current Ver"].isnull().sum()
```

```python
df.dropna(subset=["Android Ver"],inplace=True)
df["Android Ver"].isnull().sum()
```

```python
info_data()
```

```python
df["Reviews"]=df.Reviews.astype(int)
```

```python
info_data()
```

```python
df["Size"]=df.Size.apply(lambda x:x.strip("+"))
df["Size"]=df.Size.apply(lambda x:x.replace(',',''))
df["Size"]=df.Size.apply(lambda x:x.replace(",",""))
df["Size"]=df.Size.apply(lambda x:x.replace("M","e+6"))
df["Size"]=df.Size.apply(lambda x:x.replace("k",'e+3'))
df["Size"]=df.Size.replace('Varies with device',np.NaN)
```

```python
df["Size"]=pd.to_numeric(df["Size"])
```

```python
df["Price"]=df.Price.apply(lambda x:x.strip("$"))
df["Installs"]=df.Installs.apply(lambda x:x.strip("+"))
df["Installs"]=df.Installs.apply(lambda x:x.replace(',',''))
df["Installs"]=df.Installs.apply(lambda x:x.replace(",",""))
df["Price"]=pd.to_numeric(df["Price"])
df["Installs"]=pd.to_numeric(df["Installs"])
```

```python
def top10Apps(str):
    str=str.upper()
    top10=df[df["Category"]==str]
    top10app=top10.sort_values(by="Installs",ascending=False).head(10)
    
    
    fig=px.bar(x=top10app.App,y=top10app.Installs,title="Top10 Apps Installed According to Category",height=550,width=600)
    
    
    fig.update_xaxes(showgrid=False)

    fig.update_yaxes(showgrid=False, categoryorder='total ascending', ticksuffix=' ', showline=False)

    fig.update_traces(hovertemplate=None, marker=dict(line=dict(width=0)))

    fig.update_layout(margin=dict(t=80, b=0, l=70, r=40),hovermode="y unified",
                      xaxis_title=' ', yaxis_title=" ", height=450,plot_bgcolor='#333', paper_bgcolor='#333',
                      title_font=dict(size=25, color='#8a8d93', family="Lato, sans-serif"),
                      font=dict(color='#8a8d93'),
                      legend=dict(orientation="h", yanchor="bottom", y=1, xanchor="right", x=0.5),
                      hoverlabel=dict(bgcolor="black", font_size=13, font_family="Lato, sans-serif"))
    fig.show()             
```
