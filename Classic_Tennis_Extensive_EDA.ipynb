```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Skip Warnings
import warnings
warnings.filterwarnings('ignore')
warnings.filterwarnings(action = 'ignore', category = DeprecationWarning)
warnings.filterwarnings(action = 'ignore', category = FutureWarning)
```

```
df = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataVisualization/master/dataset/ClassicTennis/tenis.csv")
df.head()
```

```
df.isna().sum()
```

```
df.describe(include="all")
```

```
num_cols = ["temperature", "humidity"]
cat_cols = ["outlook", "windy","play"]
```

```
# Numerical features
df[num_cols] = df[num_cols].astype("int")

# Categorical Features
df[cat_cols] = df[cat_cols].astype("category")
for col in cat_cols:
    print(col,"--->>>",df[col].unique())
```

```
fig = plt.figure(figsize=(20,20))

for i in range(0,len(cat_cols)):
    ax = fig.add_subplot(len(cat_cols), 2, i+1)
    g = sns.countplot(df[cat_cols[i]])
    
```

```
plt.figure(figsize = (12,5))
plt.subplot(1,2,1)
sns.countplot(df["outlook"], hue = df["play"])
plt.subplot(1,2,2)
sns.countplot(df["windy"], hue = df["play"])
```

```
fig = plt.figure(figsize=(20,20))

for i in range(0,len(num_cols)):
    ax = fig.add_subplot(len(cat_cols), 2, i+1)
    g = sns.distplot(df[num_cols[i]])
    plt.grid()
```
