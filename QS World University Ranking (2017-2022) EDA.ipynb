```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
qsDF = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/master/dataset/QS%20World%20University%20Rankings%202017%20-%202022/qs-world-university-rankings-2017-to-2022-V2.csv")
```

```python
qsDF.head()
```

```python
qsDF.info()
```

```python
qsDF.drop(['link','logo'], axis=1, inplace=True)
qsDF
```

```python
round(qsDF.isnull().sum())
```

```python
round(qsDF.isnull().sum() * 100 / len(qsDF))
```

```python
sns.displot(
    data=qsDF.isnull().melt(value_name="missing"),
    y='variable', hue='missing',multiple='fill',aspect=2,palette='coolwarm')
plt.title('Missing Data Visualization')
```

```python
qsDF.drop(['score'], axis=1, inplace=True)
qsDF
```

```python
qsDF.info()
```

```python
import re
```

```python
qsDF['faculty_count'] = qsDF.faculty_count.str.replace(r'\W+','').astype(float) #to replace both '.' & ','
qsDF['international_students'] = qsDF.international_students.str.replace(r'\W+','').astype(float)
```

```python
qsDF.info()
```

```python
sns.heatmap(qsDF.corr(),cmap='inferno',annot=True)
plt.title('qsDF Correlation Matrix')
```

```python
sns.set_style('white')
sns.lmplot(data=qsDF,x='international_students',y='faculty_count',
          aspect=2,height=8)
plt.title('Plot of the linear regression between number of faculty and number of  international students', fontsize=20,
          color="green")
plt.xlabel('Number of International Students')
plt.ylabel('Number of Faculty')
```

```python
graph=qsDF[['international_students','country']].groupby('country').sum().sort_values(by='international_students', ascending=False).head(50)
graph
```

```python
sns.set_style('dark')
plt.figure(figsize=(10,12))
sns.barplot(
    data=graph, x='international_students', y=graph.index, palette='viridis')
plt.title('Top 50 Countries of Choice for International Students for the Years 2017-2022', fontsize=15,
          color="purple")
plt.ylabel('Countries', fontsize=15)
plt.xlabel('count', fontsize=15)
```
