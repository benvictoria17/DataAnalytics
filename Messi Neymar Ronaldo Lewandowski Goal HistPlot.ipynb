```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
```

```python
train0=pd.read_csv('https://raw.githubusercontent.com/benvictoria17/DataAnalytics/main/dataset/Messi%2C%20Neymar%2C%20Ronaldo%2C%20Lewandowski%20All%20Goals/messi.csv')
train1=pd.read_csv('https://raw.githubusercontent.com/benvictoria17/DataAnalytics/main/dataset/Messi%2C%20Neymar%2C%20Ronaldo%2C%20Lewandowski%20All%20Goals/neymar.csv')
train2=pd.read_csv('https://raw.githubusercontent.com/benvictoria17/DataAnalytics/main/dataset/Messi%2C%20Neymar%2C%20Ronaldo%2C%20Lewandowski%20All%20Goals/ronaldo.csv')
train3=pd.read_csv('https://raw.githubusercontent.com/benvictoria17/DataAnalytics/main/dataset/Messi%2C%20Neymar%2C%20Ronaldo%2C%20Lewandowski%20All%20Goals/lewandowski.csv')
```

```python
display(train0[0:3])
train0['count']=1
train1['count']=1
train2['count']=1
train3['count']=1
```

```python
cols=train0.columns.tolist()
print(cols)
print(len(cols))
```

```python
for i in range(12):
    data=train0.groupby(cols[i],as_index=False)['count'].sum().sort_values('count',ascending=False)
    plt.figure(figsize=(15,6))
    plt.title(cols[i])
    g = sns.barplot(x=data[cols[i]],y=data['count'],color="#4F81BD",data=data)
    g.tick_params(axis='x', labelrotation=90)
    plt.show()
```

```python
for i in range(12):
    data=train1.groupby(cols[i],as_index=False)['count'].sum().sort_values('count',ascending=False)
    plt.figure(figsize=(15,6))
    plt.title(cols[i])
    g = sns.barplot(x=data[cols[i]],y=data['count'],color="#4F81BD",data=data)
    g.tick_params(axis='x', labelrotation=90)
    plt.show()
```
