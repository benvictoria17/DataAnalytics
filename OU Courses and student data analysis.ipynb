```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 

vle = pd.read_csv("../input/open-university-learning-analytics-dataset/vle.csv")
assessments = pd.read_csv("../input/open-university-learning-analytics-dataset/assessments.csv")
courses = pd.read_csv("../input/open-university-learning-analytics-dataset/courses.csv")
studentRegistration = pd.read_csv('../input/open-university-learning-analytics-dataset/studentRegistration.csv')
studentInfo =   pd.read_csv("../input/open-university-learning-analytics-dataset/studentInfo.csv")
studentVle = pd.read_csv("../input/open-university-learning-analytics-dataset/studentRegistration.csv")
```

```python
def tabel(colname, rowname, data):
    tab = pd.crosstab(data[colname], data[rowname]) 
    tab.reset_index(level=0, inplace=True)
    
    tab['total'] = 0
    cols = [colname]
    for col in data[rowname].unique():
        tab['total'] = tab['total'] + tab[col]    
        cols.append(col)
    cols.append('total')
    for col in data[rowname].unique():
        ratiocol = col + '_%'
        tab[ratiocol] = round(100 * tab[col] / tab['total'], 2)
    return tab
```

```python
colname = 'age_band'
tabel(colname, 'final_result', studentInfo)
```
