```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
qs_17 = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/main/dataset/QS%20World%20University%20Rankings%202017-2022/QS%20World%20University%20Rankings%202017.csv")
qs_18 = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/main/dataset/QS%20World%20University%20Rankings%202017-2022/QS%20World%20University%20Rankings%202018.csv")
qs_19 = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/main/dataset/QS%20World%20University%20Rankings%202017-2022/QS%20World%20University%20Rankings%202019.csv")
qs_20 = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/main/dataset/QS%20World%20University%20Rankings%202017-2022/QS%20World%20University%20Rankings%202020.csv")
qs_21 = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/main/dataset/QS%20World%20University%20Rankings%202017-2022/QS%20World%20University%20Rankings%202021.csv")
qs_22 = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/main/dataset/QS%20World%20University%20Rankings%202017-2022/QS%20World%20University%20Rankings%202022.csv")
```

```python
data = {"df_17": qs_17, "df_18" : qs_18,"df_19": qs_19, "df_20" : qs_20,"df_21": qs_21, "df_22" : qs_22}
list(data.keys())
```

```python
years = list(data.keys())
y_ = ["2017", "2018", "2019", "2020", "2021", "2022"]
for year, y in zip(years, y_):
    sns.countplot(x = "region", data = data[year]).set_title("Count Plot by Region in " + y)
    plt.show()
```

```python
for year, y in zip(years, y_):
    plt.figure(figsize=(30,4))
    df = data[year].groupby(["country"]).agg("count").sort_values(by=["logo"], ascending = False).reset_index()
    sns.barplot(x = "country", y = "logo", data = df[df["logo"]>= 10]).set_title("Bar Plot by Country in " + y)
    plt.show()
```
