```python
import pandas as pd
pd.set_option('display.max_rows', 50)
pd.set_option('display.max_columns', 50)
pd.set_option('display.width', 1000)
```

```python
mvps = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/master/dataset/1991-2021%20NBA%20Stats/mvps.csv", encoding="latin-1", delimiter=";")
mvps
```

```python
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
mvps.head()
```

```python
players = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/master/dataset/1991-2021%20NBA%20Stats/players.csv", encoding="latin-1", delimiter=";")
players
```

```python
del players["Rk"]
```

```python
players["Player"] = players["Player"].str.replace("*","", regex=False)
players.tail(10)
```

```python
players[players["Player"] == "Delon Wright"]
```

```python
def single_team(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"]=="TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_team)
players.tail(10)
```

```python
players.index = players.index.droplevel()
players.index = players.index.droplevel()
players.tail(10)
```

```python
players[players["Player"] == "Delon Wright"]
```

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
combined[combined["Pts Won"] > 0]
combined
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
combined
```

```python
teams = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/master/dataset/1991-2021%20NBA%20Stats/teams.csv", encoding="latin-1", delimiter=";")
teams.head(20)
```

```python
teams = teams[~teams["W"].str.contains("Division")].copy()
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
sorted(teams["Team"].unique())
```

```python
sorted(combined["Tm"].unique())
```
