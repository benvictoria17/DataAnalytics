```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

lot_data = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/main/dataset/Mega%20Millions%20Lottery%20Data/mega_millions_data_2002_2022.csv")
lot_data.columns=lot_data.columns.str.replace(' ','')
lot_data['DrawDate'] = pd.to_datetime(lot_data['DrawDate'])

lot_data
```

```python
lot_data = lot_data[lot_data.DrawDate.between('2017-10-30','2022-7-22')]
lot_data
```

```python
a = lot_data.A.value_counts().head(10)

a.plot.bar(title='Most Occuring Numbers for 1st Ball Position',xlabel='Ball', ylabel='Count')
plt.xticks(rotation=1, horizontalalignment='center')

ax1 = lot_data.plot.scatter(x='A',y='DrawDate', title='Scatter plot for Ball 1 Over Time')
```

```python
t_list_a = tues_rows['A'].tolist()
t_list_b = tues_rows['B'].tolist()
t_list_c = tues_rows['C'].tolist()
t_list_d = tues_rows['D'].tolist()
t_list_e = tues_rows['E'].tolist()
t_list_mb = tues_rows['MegaBall'].tolist()
f_list_a = fri_rows['A'].tolist()
f_list_b = fri_rows['B'].tolist()
f_list_c = fri_rows['C'].tolist()
f_list_d = fri_rows['D'].tolist()
f_list_e = fri_rows['E'].tolist()

len(f_list_e)
```

```python
t_list_e[:10]
```

```python
new_a = []
new_b = []
new_c = []
new_d = []
new_e = []
new_a1 = []
new_b1 = []
new_c1 = []
new_d1 = []
new_e1 = []
def find_num(list):
    if list == t_list_a:
        for i in range(1,71):
            if i not in (list):
                new_a.append(i)
    elif list == t_list_b:
        for i in range(1,71):
            if i not in (list):
                new_b.append(i)
    elif list == t_list_c:
        for i in range(1,71):
            if i not in (list):
                new_c.append(i)
    elif list == t_list_d:
        for i in range(1,71):
            if i not in (list):
                new_d.append(i)
    elif list == t_list_e:
        for i in range(1,71):
            if i not in (list):
                new_e.append(i)
    if list == f_list_a:
        for i in range(1,71):
            if i not in (list):
                new_a1.append(i)
    if list == f_list_b:
        for i in range(1,71):
            if i not in (list):
                new_b1.append(i)
    if list == f_list_c:
        for i in range(1,71):
            if i not in (list):
                new_c1.append(i)
    if list == f_list_d:
        for i in range(1,71):
            if i not in (list):
                new_d1.append(i)
    if list == f_list_e:
        for i in range(1,71):
            if i not in (list):
                new_e1.append(i)

find_num(t_list_a)
find_num(t_list_b)
find_num(t_list_c)
find_num(t_list_d)
find_num(t_list_e)
find_num(f_list_a)
find_num(f_list_b)
find_num(f_list_c)
find_num(f_list_d)
find_num(f_list_e)
new_a1
```

```python
t = {
    'Ball 1' : new_a,
    'Ball 2' : new_b,
    'Ball 3' : new_c,
    'Ball 4' : new_d,
    'Ball 5' : new_e
}

f = {
    'Ball 1' : new_a1,
    'Ball 2' : new_b1,
    'Ball 3' : new_c1,
    'Ball 4' : new_d1,
    'Ball 5' : new_e1
}
newdf_t = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in t.items() ]))
newdf_f = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in f.items() ]))

newdf_t.head()
```
