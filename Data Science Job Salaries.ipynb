```python
df=df.iloc[:,1:]
df.head()
```

```python
import matplotlib.pyplot as plt
for i in df.columns:
    if i!='salary'and i!='job_title':
        peport_sin=df.groupby(['job_title',i]).agg({'salary':'mean'}).sort_values('salary',ascending=False).head(20)
        peport_sin.plot(kind='bar',figsize = (12,8), grid = True, color = 'teal')
        plt.title('Top 20 job title and '+i)
        plt.xlabel(None)
        plt.ylabel('Salary mean value:')
```
