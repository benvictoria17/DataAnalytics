```
plt.figure(figsize=(16,9))
top_15_uni_2017=data['university'][:15]
chart=data.groupby(top_15_uni_2017)['score'].sum().sort_values(ascending= True).plot(
kind='bar', color='r')
chart.set_xticklabels(chart.get_xticklabels(), rotation=45)
plt.title('Top 15 Universities in 2017', fontsize = 15)
plt.xlabel('university', fontsize=12)
plt.ylabel('score', fontsize=12)
plt.show()
```
