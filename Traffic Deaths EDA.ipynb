```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
sns.set_palette('viridis')
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/master/dataset/Road%20traffic%20deaths%2C%201990%20to%202019/output.csv")
```

```python
def find_minmax(x):
    min_index = df[x].idxmin()
    high_index = df[x].idxmax()
    high = pd.DataFrame(df.iloc[high_index,:])
    low = pd.DataFrame(df.iloc[min_index,:])
    print('Year with Highest ' + x + ':', df['Year'][high_index])
    print('Year with Lowest ' + x + ':', df['Year'][min_index])
    return pd.concat([high,low], axis = 1)

find_minmax('Deaths')
```
