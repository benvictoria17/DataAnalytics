```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/master/dataset/Pokemon%20with%20stats/Pokemon.csv")
df.head(20)
```

```python
df.info()
```

```python
df2 = df[ (df['Type 1']=="Fire") | (df['Type 1']=="Water") ]
df2.info()
```

```python
sns.relplot(x="Attack", y="Defense", hue= "Type 1",  data=df2, palette=["C1", "C0"] )  
```

```python
ax = sns.boxplot(x="Type 1", y="Speed", data= df)
ax = ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
```
