```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/main/dataset/Top%20100%20Rotten%20Tomatoes%20movies%20by%20genres/top_100_movies_by_genres.csv")
df.head()
```

```python
df.describe()
```

```python
genre_list = df['Genre'].unique()
print(f'There are {len(genre_list)} unique genres in the data set')
```

```python
df.dropna(inplace=True)
```

```python
avg_num_reviews = df[['Genre', 'No. of Reviews']].groupby('Genre').mean()
avg_num_reviews.plot.bar(title='Average number of reviews by Genre')
```

```python
df['RatingTomatometer'] = df['RatingTomatometer'].apply(lambda value: (float(value.strip('%'))/100))
df.head()
```

```python
avg_tomatoemeter_rating = df[['Genre', 'RatingTomatometer']].groupby('Genre').mean()
avg_tomatoemeter_rating.plot.bar(title='Average number of reviews by Rating Tomatometer')
```

```python
genre_count = df.value_counts('Genre')
genre_count.plot.bar(title='Genre Count')
```

```python
import re

df['Year'] = df['Title'].apply(lambda value: int(re.search('\d+', value).group()))
df = df[df['Year']>1000]
df.head()
```

```python
filtered_df = df[df['Year'] > 2000]
avg_reviews_by_year = filtered_df[['Year', 'No. of Reviews']].groupby('Year').mean()
avg_reviews_by_year.plot(kind='bar', title='Reviews by Year')
```

```python
avg_reviews_by_year = filtered_df[['Year', 'RatingTomatometer']].groupby('Year').mean()
avg_reviews_by_year.plot(kind='bar', title='Reviews by Year')
```
