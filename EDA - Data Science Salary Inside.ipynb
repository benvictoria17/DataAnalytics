```python
top_company= new_df["company"].value_counts()[0:15].reset_index()
top_company.columns = ["company", "Count"]
top_company
```

```python
plt.figure(figsize=(14,7))
plt.xticks(rotation=90)
plt.xlabel("Company Name")
plt.title("TOP 15 HIRING COMAPNY")
sns.barplot(x='company',y='Count',data=top_company)
plt.show()
```

```python
top_job = new_df["title"].value_counts()[0:15].reset_index()
top_job.columns = ["Job", "Count"]
top_job
```

```python
plt.figure(figsize=(14,7))
plt.xticks(rotation=75)
plt.xlabel("JOBS")
plt.title("TOP 15 JOBS")
sns.barplot(x='Job',y='Count',data=top_job)
plt.show()
```
