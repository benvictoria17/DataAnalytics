```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
imdb_initial_df = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/master/dataset/IMDB%20Movies%20Dataset/imdb_top_1000.csv")
imdb_initial_df.head()
```

```python
imdb_initial_df.info()
```

```python
imdb_initial_df.dropna(axis = 0, inplace= True)
imdb_initial_df.info()
```

```python
imdb_initial_df['Released_Year'].unique()
```

```python
imdb_initial_df[imdb_initial_df['Released_Year'] == 'PG']
```

```python
imdb_initial_df['Released_Year'].replace('PG','1995',inplace=True)
imdb_initial_df['Released_Year'] = imdb_initial_df['Released_Year'].astype('int64')
```

```python
imdb_initial_df.head(1)
```

```python
imdb_initial_df.Runtime = imdb_initial_df.Runtime.apply(lambda time : time.split()[0])
imdb_initial_df.rename(columns={'Runtime':'Runtime(min)'},inplace=True)

imdb_initial_df.Gross = imdb_initial_df.Gross.apply(lambda revenue : revenue.replace(',',''))
imdb_initial_df.Gross  = imdb_initial_df.Gross.astype('int64')
imdb_initial_df.info()
```

```python
imdb_initial_df.head(2)
```

```python
director_performance_df = imdb_initial_df.groupby('Director').Gross.agg([len,np.mean,min,max,sum])
director_performance_df.head()
```

```python
director_performance_df.sort_values('mean',ascending=False).head(5)
```

```python
imdb_initial_df.groupby('Director').IMDB_Rating.agg([len,max,min,np.mean]).sort_values('mean',ascending = False).head(5)
```

```python
imdb_initial_df_best_movies = imdb_initial_df[(imdb_initial_df['IMDB_Rating'] >=9.0)]

imdb_initial_df_best_movies.sort_values('No_of_Votes',ascending=False)
```

```python
imdb_initial_df_best_movies.sort_values('No_of_Votes',ascending=False).groupby('Director').count()
```

```python
import seaborn as sns
plt.style.use('ggplot')
plt.figure(figsize=(30,20))

plt.xticks(rotation = 90)

plt.bar(x = imdb_initial_df[imdb_initial_df['IMDB_Rating']>=8.0]['Genre'],height= imdb_initial_df[imdb_initial_df['IMDB_Rating']>=8.0]['Gross'])
plt.xlabel('Genre')
plt.ylabel('Gross Revenue')
plt.legend()
```

```python
print(imdb_initial_df.corr(method='pearson'))

sns.heatmap(imdb_initial_df.corr(method='pearson'))
```
