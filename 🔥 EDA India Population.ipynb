```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/master/dataset/Population%20of%20India%20(2050-1955)/india_population.csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.set_index('Year', inplace = True)
```

```python
df['Population'].sort_values(ascending = True).plot(kind='bar', rot = 45, color = 'lightseagreen', 
            figsize = (12,8), grid = True, fontsize = 15)
plt.title('India Population sinse 1955', fontsize = 15)
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Popualtion in BLRD', fontsize = 15)
plt.show()
```

```python
df[['World Population','Population']].sort_values('Population',ascending = True).plot(kind='bar', rot = 45, color = ['tomato','lightseagreen'], 
            figsize = (12,8), fontsize = 15)
plt.title('Worls and India population comparesing', fontsize = 15)
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Popualtion in BLRD', fontsize = 15)
plt.show()
```

```python
df['percent'] = df['Population']/df['World Population']*100
df['percent'].plot(kind='bar', rot = 45, color = 'lightseagreen', 
            figsize = (12,8), fontsize = 15)
plt.title('India population of World one in percents', fontsize = 15)
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Percent', fontsize = 15)
plt.show()
```

```python
df[['Population','Urban Population']].sort_values('Population', ascending = True).plot.bar(stacked=True, color = ['tomato','lightseagreen'],
        figsize = (12,8), grid = True, fontsize = 15)
plt.title('Urban population VS country population', fontsize = 15)
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Population in BLRD', fontsize = 15)
plt.legend(['Urban Population','Population'])
plt.show()
```

```python
df['positive'] = df['Migrants (net)'] < 0
df['Migrants (net)'].plot(kind = 'bar', 
            color = df.positive.map({True: 'tomato', False: 'lightseagreen'}), grid = True,
            figsize = (12,8), fontsize = 15)
plt.title('Indian Migrant Dinamic', fontsize = 15)
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Migrant Amount', fontsize = 15)
plt.show()
```

```python
df['Yearly Change'].sort_values(ascending = True).plot(kind = 'bar', 
            color = ['lightseagreen'], grid = True,
            figsize = (12,8))
plt.show()
```

```python
df1.set_index('Year', inplace = True)
df1.head()
```

```python
forecast_df = df.append(df1)
```

```python
forecast_df['positive'] = forecast_df.index < 2025
forecast_df['Population'].sort_values(ascending = True).plot(kind='bar', rot = 45, 
            color = forecast_df.positive.map({True:'lightseagreen',False:'Tomato'}),
            figsize = (12,8), grid = True)
plt.show()
```

```python
forecast_df['positive'] = forecast_df['Migrants (net)'] > 0
forecast_df['Migrants (net)'].plot(kind = 'bar', 
            color = forecast_df.positive.map({True: 'lightseagreen', False: 'tomato'}), grid = True,
            figsize = (12,8))
plt.show()
```
