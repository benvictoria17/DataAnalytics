```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
import plotly.express as px
import plotly.graph_objects as go
import plotly
```

```python
df1= pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/main/dataset/Bengali%20Medical%20Dataset/Bengali%20Medical%20Named%20Entity%20Recognition.csv", low_memory=False)
print(df1.shape)
df1.head().style.set_properties(**{'background-color':'Aquamarine',
                                     'color': 'purple'})
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/main/dataset/Bengali%20Medical%20Dataset/Specialist%20Classification.csv", low_memory=False)
print(df.shape)
df.head().style.set_properties(**{'background-color':'darksalmon',
                                     'color': 'purple'})
```

```python
df.isnull().sum()
```

```python
sns.set(style="ticks", context="talk", font_scale=1.5)
plt.style.use("dark_background")
```

```python
df["Gender"].value_counts()
```

```python
df["Specialist"].value_counts()
```

```python
df["Specialist"].value_counts().plot.barh(color='blue', title='Specialists',);
```

```python
color = plt.cm.RdBu(np.linspace(0,1,20))
df["Specialist"].value_counts().sort_values(ascending=False).head(10).plot.pie(y="Gender",colors=color,autopct="%0.1f%%")
plt.title("Specialists")
plt.axis("off")
plt.show()

color = plt.cm.summer(np.linspace(0,1,20))
df1["Gazetteers"].value_counts().sort_values(ascending=False).head(20).plot.pie(y="Tag",colors=color,autopct="%0.1f%%")
plt.title("Gazetteers")
plt.axis("off")
plt.show();

df["Gender"].value_counts().nlargest(4).plot.bar(color=['red','blue'], title='Gender');
```
