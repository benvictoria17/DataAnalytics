```
df=pd.read_csv('https://raw.githubusercontent.com/benvictoria17/DataAnalytics/master/dataset/Military%20power%20by%20country%202022/global%20firepower%202022%20wide.csv')
```

```
plt.figure(figsize = (8,8))
sns.barplot(y='country', x='Defense Budget',data=df.sort_values('Defense Budget', ascending=False).head(20))
```

```
plt.figure(figsize = (6,6))
sns.scatterplot(x='Total Population', y='Defense Budget', hue='country',data=df.sort_values('Defense Budget', ascending=False).head(20))
plt.legend(loc = 2, bbox_to_anchor = (1,1))
```

```
plt.figure(figsize = (6,6))
sns.scatterplot(x='External Debt', y='Defense Budget', hue='country',data=df.sort_values('Defense Budget', ascending=False).head(20))
plt.legend(loc = 2, bbox_to_anchor = (1,1))
```

```
plt.figure(figsize=(20, 10))
sns.heatmap(df.corr(),annot=True, fmt="1.1f")
```

```
df1=df.rank(ascending=False,method='min')
df1['country']=df['country']
df1['country_code']=df['country_code']
df1.head()
```

```
df2=df1.sort_values('Defense Budget', ascending=True).head(20)
df2=df2.set_index('country')
df2=df2.T
df2=df2.drop('country_code', axis=0)
df2
```

```
df2.iloc[:,0:5].plot(figsize=(30,10))
plt.xticks(rotation=90)
```
