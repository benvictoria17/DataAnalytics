```python
df1["Diabetes_binary"].value_counts()
```

```python
plt.pie(df1["Diabetes_binary"].value_counts(),labels = df1["Diabetes_binary"].unique(),autopct = "%.2f%%",shadow=True,explode=(0,0.3),radius=1.5);
plt.title("DIABETES-PATIENT OR NOT")
plt.legend()
plt.show()
```

```python
df1.groupby("Diabetes_binary")["HighBP"].value_counts()
```

```python
sns.countplot(x="Diabetes_binary", data=df1,hue="HighBP",lw=4,ec="black",hatch="/",color="r");


df1[(df1["Diabetes_binary"]==1) & (df1["BMI"]<18)]["Diabetes_binary"].value_counts()
```

```python
df1[(df1["Diabetes_binary"]==1) & (df1["BMI"]>=18) & (df1["BMI"]<25)]["Diabetes_binary"].value_counts()
```

```python
df1[(df1["Diabetes_binary"]==1) & (df1["BMI"]>=25)]["Diabetes_binary"].value_counts()
```

```python
plt.figure(figsize=(25, 18))

sns.countplot(df1.BMI[df1.Diabetes_binary == 0], color="r", label="No Diabetic");
sns.countplot(df1.BMI[df1.Diabetes_binary == 1], color="g", label="Diabetic");
plt.title("Relation between BMI and Diabetes")

plt.legend();
```

```python
print(df1[(df1["Diabetes_binary"]==1) & (df1["Smoker"]==1)]["Diabetes_binary"].value_counts())
print(df1[(df1["Diabetes_binary"]==1) & (df1["Smoker"]==0)]["Diabetes_binary"].value_counts())
```

```python
sns.countplot(x="Diabetes_binary", data=df1,hue="Smoker",lw=4,ec="black",hatch="/",color="g");
plt.title("Relation between Diabetes and Smoker")
plt.legend();

df1.groupby(["Diabetes_binary","Smoker"])["HvyAlcoholConsump"].value_counts()
```
