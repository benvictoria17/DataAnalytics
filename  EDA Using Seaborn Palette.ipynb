```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import missingno as msn 
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/benvictoria21/pandas/master/dataset/BankChurners.csv')
data.drop(['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1'
           , 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2']
           , axis = 1, inplace = True)
data.head()
```

```python
msn.matrix(data)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.describe()
```

```python
data['Attrition_Flag'].replace({'Existing Customer' : 0, 'Attrited Customer' : 1}, inplace = True)
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
plt.figure(figsize = (7, 4))
#plt.hist(data['Customer_Age'])
sns.distplot(data['Customer_Age'], color = 'm')
plt.title('Customer Age Distribution',fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (6, 6))
plt.pie(data['Gender'].value_counts(), labels = ['Female', 'Male'], autopct='%1.1f%%', 
        explode = (0, 0.1), shadow = True, startangle = 90, colors = ['pink', 'cyan'])
plt.title('Proportion of Gender count', fontsize = 20)
plt.show()
```

```python
data_attrit = data[data['Attrition_Flag'] == 1]
data_m_attrit = len(data_attrit[data_attrit.Gender == 'M'])
data_f_attrit = len(data_attrit[data_attrit.Gender == 'F'])

plt.figure(figsize = (6, 6))
plt.pie([data_m_attrit, data_f_attrit], autopct = '%1.1f%%', shadow = True, colors = ['y', 'springgreen'], explode = (0.1, 0))
plt.title('Attrited Male and Female Proportion', fontsize = 20)
plt.legend(labels = ['Male-Attrited', 'Female-Attrited'])
plt.show()
```

```python
Edu = data['Education_Level'].value_counts().to_frame('Counts') #Value_Counts Output to Dataframe

plt.figure(figsize = (7, 7))
plt.pie(Edu['Counts'], labels = Edu.index, autopct = '%1.1f%%', explode = (0.1, 0.1, 0.1, 0, 0, 0, 0), shadow = True)
plt.title('Proportion of Education Levels', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.countplot(data_attrit['Education_Level'], palette = 'magma')
plt.title('Attrited Customers Eduaction Level', fontsize = 20)
plt.show()
```

```python
data_education = data_attrit.groupby('Education_Level')['Education_Level'].count().to_frame('Count')

plt.figure(figsize = (7, 7))
plt.pie(data_education['Count'], labels = data_education.index, autopct = '%1.1f%%', explode = (0, 0, 0.1, 0, 0, 0, 0), shadow = True)
plt.title('Attrited Customers Education Level Proportion', fontsize = 20)
plt.show()
```

```python
marit = data['Marital_Status'].value_counts().to_frame('Count')

sns.set_palette('mako')
plt.figure(figsize = (6, 6))
plt.pie(marit['Count'], labels = marit.index, autopct = '%1.1f%%', explode = (0.1, 0.1, 0.1, 0.1), shadow = True,
        startangle = 90)
plt.title('Marital Status Proportion', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (7, 5))
sns.countplot(data_attrit['Marital_Status'])
plt.title('Attrited Customers Marital Status Count', fontsize = 20)
plt.show()
```

```python
category = data['Card_Category'].value_counts().to_frame('Count')

sns.set_palette('viridis_r')
plt.figure(figsize = (6, 6))
plt.pie(category['Count'], labels = category.index, explode = (0.5, 0.1, 0.75, 1.5), shadow = True, autopct = '%1.1f%%')
plt.title('Credit Card Category Proportion', fontsize = 20)
plt.show()
```

```python
income = data['Income_Category'].value_counts().to_frame('Count')

sns.set_palette('Blues_r')
plt.figure(figsize = (8, 8))
plt.pie(income['Count'], labels = income.index, explode = (0.1, 0.1, 0.1, 0, 0, 0), autopct = '%1.1f%%', shadow = True)
plt.title('Customers Income Category Proportion', fontsize = 20)
plt.show()
```

```python
depend = data['Dependent_count'].value_counts().to_frame('Count')

sns.set_palette('cubehelix_r')
plt.figure(figsize = (7, 7))
plt.pie(depend['Count'], labels = depend.index, explode = (0.1, 0.1, 0, 0, 0, 0),shadow = True, autopct = '%1.1f%%')
plt.title('Dependent Count Proportion', fontsize = 15)
plt.show()
```

```python
attrit = data['Attrition_Flag'].value_counts().to_frame('Count')

plt.figure(figsize = (5, 5))
plt.pie(attrit['Count'], labels = ['Existing Customers', 'Attrited Customers'], explode = (0, 0.1), shadow = True, startangle = 90, 
        autopct = '%1.1f%%', colors = ['silver', 'gold'])
plt.title('Proportion on Attrited Customers', fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (5, 6))
sns.countplot(data['Attrition_Flag'], palette = 'icefire')
plt.title('Target Value Count', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (8, 6))
sns.distplot(data['Months_on_book'], color = 'gold')
plt.show()
```

```python
bins = [0, 10, 20, 30, 40, 50, 60]

sns.set_palette('coolwarm_r')
plt.figure(figsize = (8, 6))
sns.countplot(pd.cut(data['Months_on_book'], bins))
plt.title('Months on Book Interval Count', fontsize = 20)
plt.show()
```

```python
bins = [500, 2000, 4000, 6000, 8000, 10000, 15000, 50000]

plt.figure(figsize = (10, 6))
sns.countplot(pd.cut(data['Credit_Limit'], bins))
plt.title('Credit Limit Interval Count', fontsize = 20)
plt.show()
```
