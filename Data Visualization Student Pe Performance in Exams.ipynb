```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
student_data = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/main/dataset/Students%20Performance%20in%20Exams/StudentsPerformance.csv")
student_data
```

```python
free_reduced_lunches = student_data[student_data.lunch == 'free/reduced']
free_reduced_lunches
```

```python
standard_lunch_students = student_data[student_data.lunch == 'standard']
standard_lunch_students
```

```python
math_score_avg = student_data.groupby('lunch', as_index=False)['math score'].mean()
reading_score_avg = student_data.groupby('lunch', as_index=False)['reading score'].mean()
writing_score_avg = student_data.groupby('lunch', as_index=False)['writing score'].mean()
math_score_avg, reading_score_avg,writing_score_avg
```

```python
sns.set_theme(style="whitegrid")

f,ax = plt.subplots()

sns.set_color_codes("pastel")


sns.barplot(x="reading score", y="lunch", data=reading_score_avg,
            label="Reading Score", color="r")

sns.barplot(x="writing score", y="lunch", data=writing_score_avg,
            label="Writing Score", color="g")

sns.set_color_codes("muted")
sns.barplot(x="math score", y="lunch", data=math_score_avg,
            label="Math Score", color="b")

sns.set_color_codes("muted")


ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(ylabel="Students",
       xlabel="Score")
sns.despine(left=True, bottom=True)
```

```python
sns.set_theme(style="dark")

sns.relplot(x="math score", y="gender",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=5, data=student_data)
sns.relplot(x="reading score", y="gender",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=5, data=student_data)
sns.relplot(x="writing score", y="gender",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=5, data=student_data)
```
