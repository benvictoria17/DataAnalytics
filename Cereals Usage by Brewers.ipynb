```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objs as go
import plotly.offline as py
import plotly.express as px
```

```python
nRowsRead = 1000 
df = pd.read_csv('https://raw.githubusercontent.com/benvictoria17/DataAnalytics/main/dataset/Cereals%20Usage%20by%20Brewers/brewers.csv', delimiter=';', encoding = "ISO-8859-1", nrows = nRowsRead)
df.dataframeName = 'brewers.csv'
nRow, nCol = df.shape
print(f'There are {nRow} rows and {nCol} columns')
df.head()
```

```python
print ("Rows     : " ,df.shape[0])
print ("Columns  : " ,df.shape[1])
print ("\nFeatures : \n" ,df.columns.tolist())
print ("\nMissing values :  ", df.isnull().sum().values.sum())
print ("\nUnique values :  \n",df.nunique())
```

```python
def plot_count(feature, title, df, size=1):
    f, ax = plt.subplots(1,1, figsize=(4*size,4))
    total = float(len(df))
    g = sns.countplot(df[feature], order = df[feature].value_counts().index[:20], palette='Set3')
    g.set_title("Number and percentage of {}".format(title))
    if(size > 2):
        plt.xticks(rotation=90, size=8)
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2.,
                height + 3,
                '{:1.2f}%'.format(100*height/total),
                ha="center") 
    plt.show()
plot_count("TOTAL_BARLEY_USAGE", "TOTAL_BARLEY_USAGE", df,4)
```

```python
plt.figure(figsize=(8,6))
fig,ax = plt.subplots(2,3,figsize=(10,8))
sns.regplot(x = 'YEAR', y = 'TOTAL_BARLEY_USAGE',data= df,color='coral',ax=ax[0][0])
sns.regplot(x = 'YEAR', y = 'BARLEY_USED_FOR_MALTING',data= df,color='coral',ax=ax[0][1])
sns.regplot(x = 'YEAR', y = 'BARLEY_USED_AS_STRAIGHT_GRAIN_FOR_BREWING_DISTILLING',data= df,color='coral',ax=ax[0][2])
sns.regplot(x = 'YEAR', y = 'TOTAL_WHEAT_USAGE',data= df,color='coral',ax=ax[1][0])
sns.countplot(x='CLOSING_STOCKS_OF_WHEAT',hue = 'YEAR', data= df,ax=ax[1][1])
sns.countplot(x='TOTAL_MAIZE_USED_AS_STRAIGHT_GRAIN_FOR_ BREWING_DISTILLING',hue = 'CLOSING_STOCKS_OF_MAIZE', data= df,ax=ax[1][2])

plt.tight_layout()
plt.show()
```
