```python
df_Gauff_Winner_sum.drop(['Total Points', 'Total Odds'], axis=1, inplace=True)
df_Gauff_Winner_sum.columns
```

```python
df_Gauff_Winner_sum = pd.get_dummies(df_Gauff_Winner_sum, columns=['Round', 'Surface_Type', 'Court'], drop_first=True)
df_Gauff_Winner_sum
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
y = df_Gauff_Winner_sum['Surface_Type_Hard']
x = df_Gauff_Winner_sum.drop('Surface_Type_Hard', axis=1)
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=42, test_size = 0.3)
lr = LogisticRegression()
model = lr.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
models = lr.fit(x_train_scaled, y_train)
models.score(x_test_scaled, y_test)
```

```python
minmax_scaler = MinMaxScaler()
x_train_minmax = minmax_scaler.fit_transform(x_train)
x_test_minmax = minmax_scaler.transform(x_test)
modelm = lr.fit(x_train_minmax, y_train)
modelm.score(x_test_minmax, y_test)
```

```python
rfc = RandomForestClassifier(n_estimators=300)
model3 = rfc.fit(x_train, y_train)
model3.score(x_test, y_test)
```

```python
model3s = rfc.fit(x_train_scaled, y_train)
model3s.score(x_test_scaled, y_test)
```

```python
model3m = rfc.fit(x_train_minmax, y_train)
model3m.score(x_test_minmax, y_test)
```

```python
svc = SVC()
model4 = svc.fit(x_train, y_train)
model4.score(x_test, y_test)
```

```python
model4s = svc.fit(x_train_scaled, y_train)
model4s.score(x_test_scaled, y_test)
```

```python
model4m = svc.fit(x_train_minmax, y_train)
model4m.score(x_test_minmax, y_test)
```
