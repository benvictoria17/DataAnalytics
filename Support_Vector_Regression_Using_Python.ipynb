```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn import svm
from mlxtend.plotting import plot_decision_regions
import numpy as np
%matplotlib inline
```

```python
eq = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/master/dataset/Sample%20Earthquake%20Data/oneyear.csv")
```

```python
eq.describe()
```

```python
eq.dropna(how='all',axis=1,inplace=True)
```

```python
f= plt.subplots(figsize=(21,21))
sn.heatmap(eq.corr(),annot=True,fmt='.1f',color='green')
```

```python
X = eq[['latitude','longitude']]
Y = [0 if elem<4.5 else 1 for elem in eq.mag]
```

```python
clf = svm.SVC()
```

```python
clf.fit(X,Y)
```

```python

plot_decision_regions(X=np.array(X), 
                      y=np.array(Y),
                      clf=clf, 
                      legend=2)

plt.xlabel(X.columns[0], size=14)
plt.ylabel(X.columns[1], size=14)
plt.title('SVM Decision Region Boundary', size=16)
```

```python
clf.predict([[69,70]])
```
