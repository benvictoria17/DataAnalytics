```python
import numpy as np 
import pandas as pd 
import scipy.stats
cereal = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/main/dataset/80%20Cereals-%20Nutrition%20data%20on%2080%20cereal%20products/cereal.csv")
```

```python
cereal.describe()
```

```python
scipy.stats.chisquare(f_obs= cereal["fat"].value_counts())
```

```python
scipy.stats.chisquare(f_obs= cereal["fiber"].value_counts())
```

```python
acrossTab = pd.crosstab( cereal["fat"], cereal["fiber"])
acrossTab
```

```python
scipy.stats.chi2_contingency(acrossTab)
```
