```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/2803Aditya/PlayStore-Dataset/master/googleplaystore.csv')
df = (df.sort_values(['Category', 'Rating'], ascending = [True, False])
      .where(df['Rating'] <= 5.0)
      .set_index('App')
      .dropna()
      .drop('Reviews', axis = 1)
      .drop_duplicates())
```

```python
df.sort_values('Installs', ascending = False).head(5)
```
