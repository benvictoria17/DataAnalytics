```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
raw_data = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/master/dataset/Financial%20Banking%20Marketing%20Campaign/Example-bank-data.csv")
raw_data.head()
```

```python
data=raw_data.copy()
data= data.drop(['Unnamed: 0'],axis=1)
data.head()
```

```python
data['y'] = data['y'].map({'yes':1, 'no':0})
data.head()
```

```python
data.describe()
```

```python
y=data['y']
x1=data['duration']
```

```python
x= sm.add_constant(x1)
reg_log= sm.Logit(y,x)
results_log = reg_log.fit()
print(results_log.summary())
```

```python
plt.scatter(x1,y,color = 'C1')
plt.xlabel('Duration', fontsize = 20)
plt.ylabel('Subscription', fontsize = 20)
plt.show()
```
