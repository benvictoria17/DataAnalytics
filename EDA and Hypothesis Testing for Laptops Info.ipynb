```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/benvictoria17/DataAnalytics/master/dataset/Laptops%20Info/Final_Dataframe.csv')
data.head()
```

```python
data.info()
```

```python
data.dropna(inplace=True)
data.drop_duplicates(keep = False, inplace=True) 
```

```python
data["brand"].unique()
```

```python
data.groupby('brand').mean().reset_index()
```

```python
data[['brand', 'laptop_name', 'disk_space', 'old_price']].sort_values(by='old_price').head(3)
```

```python
data[['brand', 'laptop_name', 'disk_space', 'old_price']].sort_values(by='old_price',ascending=False).head(3)
```

```python
def laptopbrand(x):
    y = data[['brand','laptop_name','display_size','processor_type', 'graphics_card', 'disk_space', 'discount_price', 'old_price', 'ratings_5max']][data["brand"] == x]
    y = y.sort_values(by="brand",ascending=False)
    return y
dataHP = laptopbrand('HP')
dataHP.sort_values(by='old_price',ascending=False)
```

```python
plt.figure(figsize=(7,7))
sns.distplot(dataHP['old_price'], color = 'blue', label = 'old_price')
sns.distplot([dataHP['discount_price']], color = 'green', label = 'discount_price')
plt.title('Summary Of Brand HP Based On Old Price')
plt.show()
```

```python
data['gap'] = data['old_price'] - data['discount_price']
sns.distplot(data['gap'])
plt.title('Difference Between Old Price And Discount Price')
plt.show()
```

```python
from scipy.stats import kstest
```

```python
stats, pval = kstest(data['gap'], 'norm')
stats
```

```python
pval
```

```python
pval < 0.05
```

```python
from scipy.stats import wilcoxon
```

```python
rank, pval = wilcoxon(data['gap'], alternative = 'less')
rank
```

```python
pval
```

```python
pval < 0.05
```
