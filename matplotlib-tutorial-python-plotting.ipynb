```python
# Import the `pandas` library as `pd`
import pandas as pd

# Load in the data with `read_csv()`
digits = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/optdigits/optdigits.tra",
                     header=None)

# Print out `digits`
print(digits)
```

```python
digits.describe()
```

```python
# Inspect the first 5 rows of `digits`
first = digits.head(5)

# Inspect the last 5 rows
last = digits.tail(5)
```

```python
# Take a sample of 5
digits.sample(5)
```

```python
# Identifiy missing values
pd.isnull(digits)
```

```python
# Drop rows with missing values
digits.dropna(axis=0)

# Drop columns with missing values 
digits.dropna(axis=1)
```

```python
digits.interpolate()
```

```python
# Import `PCA` from `sklearn.decomposition`
from sklearn.decomposition import PCA

# Build the model
pca = PCA(n_components=2)

# Reduce the data, output is ndarray
reduced_data = pca.fit_transform(digits)

# Inspect shape of the `reduced_data`
reduced_data.shape

# print out the reduced data
print(reduced_data)
```

```python
# Pearson correlation
digits.corr()

# Kendall Tau correlation
digits.corr('kendall')

# Spearman Rank correlation
digits.corr('spearman')
```
