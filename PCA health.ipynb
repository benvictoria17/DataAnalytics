```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.decomposition import PCA
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
df
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
```

```python
df_scaled = scaler.transform(df)
```

```python
covariance_matrix = PCA(n_components=30)
covariance_matrix.fit(df_scaled)
```

```python
variance = covariance_matrix.explained_variance_ratio_
var = np.round(variance, decimals=3)
cumulative_var = np.cumsum((var)*100)
cumulative_var
```

```python
plt.plot(cumulative_var)
plt.title('SCREE PLOT (selecting optimal features)')
plt.xlabel('No of features')
plt.ylabel('% of variance explained')
plt.show()
```

```python
pca = PCA(n_components=5)
pca.fit(df_scaled)
```

```python
df_pca = pca.transform(df_scaled)
df_scaled.shape
```

```python
df_pca.shape
plt.scatter(df_pca[:,0], df_pca[:,1], c=cancer['target'], cmap='rainbow')
plt.show()
```
